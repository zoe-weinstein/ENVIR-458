# Homework 5: Data Visualization with Matplotlib and Seaborn

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

### Problem 1: Basic Plotting with Matplotlib (1 pt)
Objective: Create a line plot using Matplotlib to visualize the trend of a given dataset.

Dataset: Use np.linspace(0, 10, 100) for the x-values and np.sin(x) for the y-values.
Requirements:
* Add a title to the plot.
* Label the x-axis as "X" and the y-axis as "sin(X)".
* Customize the line style to be dashed and the color to blue.

```python
# Creating a line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x, y, linestyle='--', color='blue')
ax.set_title('Sine Values')
ax.set_xlabel('X Values')
ax.set_ylabel('Sine Values')
```

### Problem 2: Exploring Distributions with Seaborn (1 pt)

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.histplot(tips["total_bill"], kde = True, bins = 20)
plt.title("Distribution of Total Bill")
plt.xlabel("Bill Total Amount")
plt.ylabel("Frequency")
plt.show()
```

### Problem 3: Categorical Data Analysis (2 pt)

```python
fig, ax = plt.subplots(1, 2)
sns.boxplot(tips, x = "time", y = "tip", ax=ax[0])
sns.violinplot(tips, x = "time", y = "tip", ax=ax[1])
ax[0].set_title("Box Plot")
ax[1].set_title("Violin Plot")
ax[0].set_xlabel("Time")
ax[0].set_ylabel("Tip")
ax[1].set_xlabel("Time")
plt.show()
```

### Problem 4: Multivariate Analysis with Scatter Plots (2 pt)

```python
# Problem 4: Multivariate Analysis with Scatter Plots
sns.lmplot(data=tips, x="total_bill", y="tip", hue="smoker", markers=["o", "s"], palette="coolwarm")
plt.title("Total Bill vs. Tip Colored by Smoker Status")
plt.show()
```

### Problem 5: Heatmap of Correlation Matrix


```python

# Problem 5: Heatmap of Correlation Matrix
corr_matrix = tips.corr(numeric_only=True)
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", center=0, linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

### Problem 6: Pair Plot Analysis

```python
# Problem 6: Pair Plot Analysis
sns.pairplot(tips, hue="time", diag_kind="kde", palette="muted")
plt.show()
```
